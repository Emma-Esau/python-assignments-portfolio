```python
import sympy as sp
x, y = sp.symbols('x y')
```

```python
f = sp.exp(x)*sp.sin(y)+y**3

df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)

print("df_dx =", df_dx)
print("df_dy =", df_dy)
```

```python
h = sp.ln(x**2+y**2)

d2h_dx2 = sp.diff(h, x, x)
d2h_dy2 = sp.diff(h, y, y)
d2h_dxdy = sp.diff(h, x, y)

print("d2h_dx2 = ", d2h_dx2)
print("d2h_dy2 = ", d2h_dy2)
print("d2h_dxdy =", d2h_dxdy)
```

```python
g = x**2*y+x*y**2


x_val = 1
y_val = -1

dg_dx = sp.diff(g, x)
dg_dy = sp.diff(g, y)

dg_dx_val = dg_dx.subs({x: x_val, y: y_val})
dg_dy_val = dg_dy.subs({x: x_val, y: y_val})

mag = sp.sqrt(dg_dx_val**2 + dg_dy_val**2)

print("dg_dx =", dg_dx_val)
print("dg_dy =", dg_dy_val)
print("magnitude =", mag.evalf())
```

```python
import numpy as np
import matplotlib.pyplot as plt

x, y = sp.symbols('x y')
j = x**3 - 3*x*y + y**3

j_func = sp.lambdify((x, y), j, 'numpy')

x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = j_func(X, Y)

plt.contourf(X, Y, Z, levels=50, cmap='magma')
plt.colorbar()
plt.title('Contour plot of $j(x, y) = x^3 - 3xy + y^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x, y = sp.symbols('x y')
k = x**5 - 3*y + x*y**3

k_func = sp.lambdify((x, y), k, 'numpy')

x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = k_func(X, Y)

plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $k(x, y) = x^5 - 3y + xy^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```
